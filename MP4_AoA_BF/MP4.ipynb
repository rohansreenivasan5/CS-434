## Your Implementation
You will implement three functions here. For part 1, implement `delay_and_sum_aoa(mic_data_folder, FS, MIC_OFFSETS)` and `music_aoa(mic_data_folder, FS, MIC_OFFSETS)` to get the incoming signal's [polar angle](https://en.wikipedia.org/wiki/Polar_coordinate_system). You may assume there is only one direction path. MUSIC implementation is an extra 10% credit. Please change the return value of `music_implemented` function to True if you implemented MUSIC as well.

For part 2, implement localization function `aoa_localization(aoas, mic_data_folder)`. Do **NOT** change function signatures. You are, however, free to define and use helper functions. 

You are encouraged to inspect, analyze and optimize your implementation's intermediate results using plots and outputs. Your implementation for `aoa_localization` function should **NOT** output any plots or data. It should only return the user's calculated location.

````python
import os
import math
from scipy.signal import find_peaks
from numpy.linalg import norm
from scipy.signal import correlate
from scipy.constants import speed_of_sound as c
from numpy.linalg import svd, eig
from scipy.optimize import least_squares

def music_implemented():
    return True


def music_aoa(mic_data_folder, FS, MIC_OFFSETS):
    array_locs = np.genfromtxt(os.path.join(mic_data_folder, 'config.csv'), delimiter=",")
    aoas = []

    # Define steering vector function
    def steering_vector(theta, mic_offsets, frequency=FS, speed_of_sound=c):
        wavelength = speed_of_sound / frequency
        k = 2 * np.pi / wavelength
        a = np.exp(-1j * k * np.dot(mic_offsets, [np.cos(theta), np.sin(theta)]))
        return a

    for array_index in range(len(array_locs)):
        # Load the microphone data for each array
        mic_data = pd.read_csv(os.path.join(mic_data_folder, f'{array_index}.csv'), header=None).values
        num_mics = mic_data.shape[1]
        
        # Calculate the spatial covariance matrix
        R = np.dot(mic_data.T, mic_data.conj()) / mic_data.shape[0]

        # Perform eigenvalue decomposition
        eigenvals, eigenvects = eig(R)
        
        # Sort eigenvectors by eigenvalues in descending order
        idx = eigenvals.argsort()[::-1]
        eigenvals = eigenvals[idx]
        eigenvects = eigenvects[:, idx]

        # Assume the number of sources is 1 for simplicity
        noise_subspace = eigenvects[:, 1:]

        # MUSIC scan over a grid of possible AoAs
        angles = np.linspace(-np.pi/2, np.pi/2, 360)
        music_spectrum = np.zeros(len(angles))

        for i, angle in enumerate(angles):
            sv = steering_vector(angle, MIC_OFFSETS)
            music_spectrum[i] = 1 / np.dot(sv.conj().T, np.dot(noise_subspace, noise_subspace.conj().T).dot(sv))

        # Find the angle with the maximum peak in the MUSIC spectrum
        aoas.append(angles[np.argmax(music_spectrum)])
    
    # Convert AoAs from radians to degrees andensure they are within [0, 360] range
    aoas = np.rad2deg(aoas)
    aoas = [aoa if aoa >= 0 else aoa + 360 for aoa in aoas]

    return aoas


def delay_and_sum_aoa(mic_data_folder, FS, MIC_OFFSETS):
    # Load microphone array locations from the config file
    array_locs = np.genfromtxt(os.path.join(mic_data_folder, 'config.csv'), delimiter=",")

    aoas = []

    # Define a function to calculate the delay between two microphones for a given angle
    def calculate_delay(mic1, mic2, angle, speed_of_sound=c, sampling_rate=FS):
        # Calculate the distance difference between the paths from the sound source to each microphone
        distance_diff = np.dot(mic2 - mic1, np.array([np.cos(angle), np.sin(angle)]))
        # Convert distance difference to time delay
        time_delay = distance_diff / speed_of_sound
        # Convert time delay to number of samples
        sample_delay = np.round(time_delay * sampling_rate)
        return int(sample_delay)

    # Process each microphone array separately
    for array_index in range(len(array_locs)):
        mic_signals = pd.read_csv(os.path.join(mic_data_folder, f'{array_index}.csv'), header=None).values

        # Initialize variables to store the best AoA and its corresponding summed signal strength
        best_aoa = None
        best_sum = -np.inf

        # Search for the AoA that maximizes the aligned signal sum
        for aoa_deg in np.arange(0, 360, 1):
            aoa_rad = np.deg2rad(aoa_deg)
            aligned_signals = np.zeros_like(mic_signals[:, 0])

            # Align signals based on calculated delays
            for i in range(len(MIC_OFFSETS)):
                for j in range(i + 1, len(MIC_OFFSETS)):
                    delay = calculate_delay(np.array(MIC_OFFSETS[i]), np.array(MIC_OFFSETS[j]), aoa_rad)
                    if delay >= 0:
                        aligned_signal = np.roll(mic_signals[:, j], -delay)
                    else:
                        aligned_signal = np.roll(mic_signals[:, i], delay)
                    aligned_signals += aligned_signal

            # Sum the aligned signals and check if this is the best alignment so far
            current_sum = np.sum(aligned_signals**2)
            if current_sum > best_sum:
                best_sum = current_sum
                best_aoa = aoa_deg

        aoas.append(best_aoa)

    return aoas



def aoa_localization(angle_degrees, folder_path):
    array_coords = pd.read_csv(f"{folder_path}/config.csv", header=None).values
    
    eq_components = []
    for angle, (coord_x, coord_y) in zip(angle_degrees, array_coords):
        distance = coord_x * np.cos(angle) + coord_y * np.sin(angle)
        eq_components.append((np.cos(angle), np.sin(angle), distance))
    
    eq_matrix = np.array([[np.cos(deg), np.sin(deg)] for deg in angle_degrees])
    eq_vector = np.array([coord_x * np.cos(deg) + coord_y * np.sin(deg) for deg, (coord_x, coord_y) in zip(angle_degrees, array_coords)])
    
    estimated_position, _, _, _ = np.linalg.lstsq(eq_matrix, eq_vector, rcond=None)
    
    return tuple(estimated_position)
````
